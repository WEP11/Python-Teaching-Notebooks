# Introducing MatPlotLib
In this section we will learn how to plot our data into a basic line graph. This introduction will help us learn more advanced plots later on in the workshop.

## Getting Ready
We will begin by using our data reading skills we learned previously. Using Pandas, read in the sample Oklahoma Mesonet dataset. There is missing data after 11:75, so let's subset our frame before we begin visualizing the data.

```python
import pandas as pd

# Read the CSV, setting whitespace as the separator and row 2 as header
df = pd.read_csv('20190425nrmn.mts', index_col=2, sep='\s+', header=2)

# We start to have missing data after 1175, so let's clean up our final
# product by only using data from before 1175
plotData = df.loc[:1175]
```

## Using MatPlotLib
We're going to start by importing the MatPlotLib PyPlot library. This is basic entry point for using MatPlotLib, which let's us set up some basic plots easily: 

```python
# Import MatPlotLib
import matplotlib.pyplot as plt
```

After importing the module, we can immediately begin plotting. Let's start with a very basic plot of our air temperature data. We use the __plot__ method to create a line graph.

```python
# Make a basic line plot using the TAIR values
plt.plot(plotData['TAIR'].values)
plt.show()
```

In 2 lines we were able to create a very nice, although basic, plot! But this is far from publication ready, so let's start by getting our times onto the X axis. We can do this a couple of ways:

```python
# Pass both the time values (X), and TAIR (Y) values
plt.plot(plotData.index.values, plotData['TAIR'].values, 'r-')
plt.show()
```

In this first example we are passing an X and Y parameter to the plot function, and changing the line color to red. Now let's add the 9m Temperature and make its color blue!

```python
# Arguments are (X1, Y1, plot options, X2, Y2, plot options)
plt.plot(plotData.index.values, plotData['TAIR'].values, 'r-',plotData.index.values, plotData['TA9M'].values, 'b-')
plt.show()
```

This is great, but MatPlotLib has a shortcut we could be using! It can recognize a simple 2 dimensional array of X/Y values:

```python
plt.plot(plotData['TAIR'], 'r-', plotData['TA9M'], 'b-', plotData['RELH'], 'g-')
plt.show()
```

Notice how you can keep adding more variables in the same way you added the first two? When there are many variables, this method of adding them can be unwieldy, but we will talk about ways to sequentially add variables in the next part of the workshop.

## Adding Plot Labels

Now let's keep focusing on our two temperatures, but start polishing up our nice graph! First we need to add some axis labels. We do this with very simple __xlabel__ and __ylabel__ methods.

```python
plt.plot(plotData['TAIR'], 'r-', plotData['TA9M'], 'b-')
plt.xlabel('Time (CST)')
plt.ylabel('Temperature (C)')

plt.show()
```

Adding a legend is also easy, thanks to the __legend__ method. We can easily pass the legend labels by giving a list of labels in the same order as our variables.

```python
# Set up the plot
plt.plot(plotData['TAIR'], 'r-', plotData['TA9M'], 'b-')
plt.xlabel('Time (CST)')
plt.ylabel('Temperature (C)')

# Provide labels for each line. Notice how they are in the same
# order as they are in the plot function?
plt.legend(['2M Temperature', '9M Temperature'])

plt.show()
```

The last thing our plot needs is a good title. This is done by the well named __title__ method.

```python
# Set up the plot
plt.plot(plotData['TAIR'], 'r-', plotData['TA9M'], 'b-')

plt.xlabel('Time (CST)')
plt.ylabel('Temperature (C)')

plt.legend(['2M Temperature', '9M Temperature'])

plt.title('An Incredible Weather Event in Central Oklahoma')
plt.show()
```

## Key Points
* [__matplotlib.pyplot__](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot) : This is the module we import to use MatPlotLib's basic functions
* [__plt.plot()__](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) : This is will create a basic line plot
* [__plt.show()__](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.show.html#matplotlib.pyplot.show) : This will show the plot you've created
* [__plt.xlabel()__](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xlabel.html#matplotlib.pyplot.xlabel) : This sets the x axis label
* [__plt.ylabel()__](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.ylabel.html#matplotlib.pyplot.ylabel) : This sets the y axis label
* [__plt.legend()__](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html#matplotlib.pyplot.legend) : This creates a legend for the plot
* [__plt.title()__](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.title.html#matplotlib.pyplot.title) : This creates a title for the plot
